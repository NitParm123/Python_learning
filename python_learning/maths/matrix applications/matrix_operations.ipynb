```python
#matrix multiplication
def matrix_multiplication(A, B):
    #get number of rows and cols of each matrix
    rowA = len(A)
    colA = len(A[0]) if A else 0
    rowB = len(B)
    colB = len(B[0]) if B else 0

    #check if multiplication is eligible or not
    if(colA != rowB):
        raise ValueError("Number of columns in A must be equal to number of rows in B")

    #define a result matrix
    result = [[0 for _ in range(colB)] for _ in range(rowA)]

    for i in range(rowA):
        for j in range(colB):
            for k in range(colA):
                result[i][j] += A[i][k] * B[k][j]

    return result

    
```

```python
#computing matrix exponentiation

#this is multiplication of matrix
def multiply(A, B):
    n = len(A)
    #define a result matrix
    result = [[0]*n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result


#compute power of matrix
def matrix_power(A, n):
    size = len(A)
    #define a identity matrix
    result = [[1 if i == j else 0 for j in range(size)] for i in range(size)]

    while(n > 0):
        if(n%2 == 1):
            result = multiply(result, A)
        A = multiply(A, A)
        n //= 2
    return result
    
    
    
```
